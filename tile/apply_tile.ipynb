# png tile -- cls

## train set

```python
import os
import json
from tile import Cutter

file_dir = '/media/ldy/7E1CA94545711AE6/OSCC/scaledown8_png'
file_mask_dir = '/media/ldy/7E1CA94545711AE6/OSCC/mask_x8/std_mask'
sample_type = 'cls'

meta_file = '/media/ldy/7E1CA94545711AE6/OSCC/meta/train_val.json'

with open(meta_file, 'r') as f:
    cnt = json.load(f)
```

```python
slide_list = cnt['train']
save_patch_dir = '/media/ldy/e5a10f4e-18fd-4656-80d8-055bc4078655/OSCC-Tile/cls_8x/train'

cutter = Cutter(slide_list,
                file_dir,
               file_mask_dir,
               save_patch_dir,
               sample_type=sample_type)

patch_size = 224
overlap = 28
filter_rate = 0.3

cutter.sample_and_store_patches_png(patch_size,
                                   overlap,
                                   filter_rate)
```

## val set

```python
slide_list = cnt['val']
save_patch_dir = '/media/ldy/e5a10f4e-18fd-4656-80d8-055bc4078655/OSCC-Tile/cls_8x/val'

cutter = Cutter(slide_list,
                file_dir,
                file_mask_dir,
                save_patch_dir,
                sample_type=sample_type)

patch_size = 224
overlap = 28
filter_rate = 0.3

cutter.sample_and_store_patches_png(patch_size,
                                   overlap,
                                   filter_rate)

```
